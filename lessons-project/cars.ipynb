Yapay Zeka Modeli Oluşturulması

```python
#istenen verinin elde edilmesi
y = araba_veri["price_usd"].values
```

```python
y
```

```python
y.shape
```

```python
#özellikler, geriye kalanlar
x = araba_veri.drop("price_usd",axis=1)
```

```python
x.shape
```

```python
from sklearn.ensemble import RandomForestRegressor 
regressor = RandomForestRegressor(n_estimators=10, random_state=0)
```

```python
regressor.fit(x, y)
```

```python
liste_sutun = list(x.iloc[0].keys())
print(liste_sutun)
liste = list(x.iloc[0].values)
liste1 = list(x.iloc[15].values)
liste2 = list(x.iloc[1655].values)
liste3 = list(x.iloc[6166].values)
liste4 = list(x.iloc[8000].values)
print(liste)
```

```python
y_pred = regressor.predict([liste,liste1,liste2,liste3,liste4])
print(y_pred)
```

```python
#test ve train verilerinin ayrılması
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3)
```

```python
print("x_train uzunluk : {}".format(len(x_train)))
print("x_test uzunluk : {}".format(len(x_test)))
print("y_train uzunluk : {}".format(len(y_train)))
print("y_test uzunluk : {}".format(len(y_test)))
```

```python
#verilerin kullanılabilir duruma indingenmesi
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

x_train = scaler.fit_transform(x_train)
x_test = scaler.fit_transform(x_test)
```

```python
#yapay zeka modelinin oluşturulması
import tensorflow as tf
#model oluşturulması
from tensorflow.keras.models import Sequential
#katmanlar ile ilgili işlemlerin yapılması
from tensorflow.keras.layers import Dense, Activation
#overfittin çözme
from tensorflow.keras.callbacks import EarlyStopping
```

```python
#del model
```

```python
#model oluşturulması
model = Sequential()
model.add(Dense(units=30,activation="relu"))
model.add(Dense(units=15,activation="relu"))
model.add(Dense(units=15,activation="relu"))
#model.add(Dense(units=1,activation="sigmoid"))
model.compile(loss="MeanSquaredError",optimizer = "adam")
```

```python
#model eğitimi
model.fit(x=x_train, y=y_train,validation_data=(x_test,y_test),batch_size=512 ,epochs=900)
```

```python
#model grafiği oluşturulması
kayipVerisi = pd.DataFrame(model.history.history)
kayipVerisi.plot()
```
