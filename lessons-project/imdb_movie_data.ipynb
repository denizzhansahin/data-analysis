            20-39 genellikle olumsuz,

            40-59 genellikle karışık puanları yer alır,

            60-79 genellikle olumlu puan,

            80-100 genellikle yüksek puan ve olumlu filmlerdir.

time_minute : Film süreleri ile ilişkilidir. Film sürelerini belirtir.

vote : Bir filmin kaç kişi tarafındn oylandığını belirtir.

gross_earning : Brüt hasılatı temsil eder

Veri Seti Düzenleme

```python
#movie sütunu silme
imdb_film_veri_seti = imdb_film_veri_seti.drop("movie",axis=1)
```

```python
#veri seti içinde yer alan '_ ve I değerlerine 0 ile değiştirme
imdb_film_veri_seti['gross_earning'] = imdb_film_veri_seti['gross_earning'].str.replace("-", '0').str.replace("'", '0')
imdb_film_veri_seti['year'] = imdb_film_veri_seti['year'].str.replace("I", "").str.replace("I", "")
imdb_film_veri_seti['year'] = imdb_film_veri_seti['year'].str.replace("-", '0').str.replace("'", '0')
imdb_film_veri_seti['year'] = imdb_film_veri_seti['year'].str.replace("-", '0').str.replace("'", '0')
imdb_film_veri_seti['genre.1'] = imdb_film_veri_seti['genre.1'].str.replace("-", '0').str.replace("'", '0')
imdb_film_veri_seti['genre.2'] = imdb_film_veri_seti['genre.2'].str.replace("-", '0').str.replace("'", '0')
imdb_film_veri_seti['genre.3'] = imdb_film_veri_seti['genre.3'].str.replace("-", '0').str.replace("'", '0')
imdb_film_veri_seti['metascore'] = imdb_film_veri_seti['metascore'].str.replace("-", '0').str.replace("'", '0').astype(float)
imdb_film_veri_seti['time_minute'] = imdb_film_veri_seti['time_minute'].str.replace("-", '0').str.replace("'", '0')
imdb_film_veri_seti['certificate'] = imdb_film_veri_seti['certificate'].str.replace("-", '0').str.replace("'", '0')
```

```python
#time_minute sütunu düzenleme
imdb_film_veri_seti["time_minute"] = imdb_film_veri_seti["time_minute"].str.extract('(\d+)', expand=False).astype(float)
```

```python
#gross_earning sütunu düzenleme
imdb_film_veri_seti["gross_earning"] = imdb_film_veri_seti["gross_earning"].str.replace("$", "").str.replace("M", "").str.replace("#", "").astype(float)
```

```python
#year sütunu düzeltme
imdb_film_veri_seti["year"] = imdb_film_veri_seti["year"].str.replace("(", "").str.replace(")", "").astype(int)
```

Veri Setinin Ayarlanması, IMDB Ratin Sonucu Bulunması

```python
#parçalama işlemleri

#istenen verinin elde edilmesi
y = imdb_film_veri_seti["imdb_rating"].values

#özellikler, geriye kalanlar
x = imdb_film_veri_seti.drop("imdb_rating",axis=1)

#test ve train verilerinin ayrılması
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.3)
```

```python
#parçalanan verilerin bilgileri (boyut ve uzunluk)
print("x uzunluk : {}".format(len(x)))
print("y uzunluk : {}".format(len(y)))

print("x_train uzunluk : {}".format(len(x_train)))
print("x_test uzunluk : {}".format(len(x_test)))
print("y_train uzunluk : {}".format(len(y_train)))
print("y_test uzunluk : {}".format(len(y_test)))

print("x_train shape : {}".format(x_train.shape))
print("x_test shape : {}".format(x_test.shape))
print("y_train shape : {}".format(y_train.shape))
print("y_test shape : {}".format(y_test.shape))
```

Sklearn ile Model Eğitilmesi ve Tahmin Sonuçları

```python
#kütüphane ekleme
from sklearn.ensemble import RandomForestRegressor 
regressor = RandomForestRegressor(n_estimators=10, random_state=0)
```

```python
#model eğitimi
regressor.fit(x, y)
```

```python
#model tahmini
liste_sutun = list(x.iloc[0].keys())
print(liste_sutun)
liste = list(x.iloc[0].values)
liste1 = list(x.iloc[150].values)
liste2 = list(x.iloc[300].values)
liste3 = list(x.iloc[600].values)
liste4 = list(x.iloc[950].values)
print(liste)

y_pred = regressor.predict([liste,liste1,liste2,liste3,liste4])
print(y_pred)
```

TensorFlow ile Model Eğitilmesi ve Tahmin Sonuçları

```python
#verilerin scaler edilmesi
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

x_train = scaler.fit_transform(x_train)
x_test = scaler.fit_transform(x_test)
```

```python
#kütüphane ekleme
import tensorflow as tf
#model oluşturulması
from tensorflow.keras.models import Sequential
#katmanlar ile ilgili işlemlerin yapılması
from tensorflow.keras.layers import Dense, Activation
#overfittin çözme
from tensorflow.keras.callbacks import EarlyStopping
```

```python
#model oluşturulması
model = Sequential()
model.add(Dense(units=10,activation="relu"))
model.add(Dense(units=5,activation="relu"))
model.add(Dense(units=5,activation="relu"))
#model.add(Dense(units=1,activation="sigmoid"))
model.compile(loss="MeanSquaredError",optimizer = "adam")
```

```python
#model eğitilmesi
model.fit(x=x_train, y=y_train,validation_data=(x_test,y_test),batch_size=512 ,epochs=250)
```

```python
#model grafiği oluşturulması
kayipVerisi = pd.DataFrame(model.history.history)
kayipVerisi.plot()
```

```python
#tahmin yapılması
tahmin = model.predict(x_test)
```
